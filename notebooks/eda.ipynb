```python
# Document for performing exploratory data analysis (EDA) on a dataset using pandas and matplotlib.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Load the dataset
movies = pd.read_csv(r'C:\Users\nilsm\OneDrive\Skrivbord\BIG data\ml-32m (1)\ml-32m\movies.csv')
reviews = pd.read_csv(r'C:\Users\nilsm\OneDrive\Skrivbord\BIG data\ml-32m (1)\ml-32m\ratings.csv')
tags = pd.read_csv(r'C:\Users\nilsm\OneDrive\Skrivbord\BIG data\ml-32m (1)\ml-32m\tags.csv')
links = pd.read_csv(r'C:\Users\nilsm\OneDrive\Skrivbord\BIG data\ml-32m (1)\ml-32m\links.csv')
```

```python
# Movies --------------------------------
print('Head:\n', movies.head())
print('\nInfo:')
movies.info()
print('\nDescribe:\n', movies.describe())
```

```python
# Reviews --------------------------------
print('Head:\n', reviews.head())
print('\nInfo:')
reviews.info()
print('\nDescribe:\n', reviews.describe())
```

```python
# Tags --------------------------------
print('Head:\n', tags.head())       
print('\nInfo:')
tags.info()
print('\nDescribe:\n', tags.describe())
```

```python
# Links --------------------------------
print('Head:\n', links.head())
print('\nInfo:')
links.info()
print('\nDescribe:\n', links.describe())
```

```python
# Further analysis and visualizations

# Histogram of movie ratings
plt.figure(figsize=(8, 2.5)) 
plt.hist(reviews['rating'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Frequencies of unique tags
tag_frequencies = tags['tag'].value_counts()
print('Frequencies of unique tags:\n', tag_frequencies)

# Histogram of tag frequencies
plt.figure(figsize=(8, 2.5))
plt.bar(range(3000), tag_frequencies.values[:3000], color='green', alpha=0.7, label='Tag Frequencies', width=1)
plt.xticks([i for i in np.arange(0, 3000, 500)])
plt.title('Distribution of Tag Frequencies')
plt.xlabel('Tag')
plt.ylabel('Count')
plt.ylim(0, max(tag_frequencies.values[:3000]) + 100)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Most popular movies based on average rating
average_ratings = reviews.groupby('movieId')['rating'].mean().sort_values(ascending=False)
average_ratings = pd.merge(average_ratings, movies[['movieId', 'title']], on='movieId')
print('Top 10 movies by average rating:\n', average_ratings.head(10))
```

```python
# Most reviewed movies
review_counts = reviews['movieId'].value_counts()
review_counts = pd.merge(review_counts.rename('review_count'), movies[['movieId', 'title']], left_index=True, right_on='movieId')
print('Top 10 most reviewed movies:\n', review_counts.head(10))
```

```python
# Frequency of movies per genre
# Split genres and explode the DataFrame
genre_counts = movies['genres'].str.get_dummies(sep='|').sum().sort_values(ascending=False)
# Bar chart of movie counts per genre
plt.figure(figsize=(10, 2.5))
plt.bar(genre_counts.index, genre_counts.values, color='purple', alpha=0.7)
plt.title('Movie Counts per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Most reviewed genres (with combinations of genres allowed)
# Merge reviews with movies to get genres
most_reviewed_genres = pd.merge(reviews, movies[['movieId', 'genres']], on='movieId')
most_reviewed_genres = most_reviewed_genres.explode('genres')
most_reviewed_genres = most_reviewed_genres['genres'].value_counts()
print('Top 10 most reviewed genres:\n', most_reviewed_genres.head(10))
print('Movies can have multiple genres, so the total count exceeds the number of unique genres.')
```
